```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
file='Employee.csv'
df=pd.read_csv(file)
print(df.head())
```

```python
sns.histplot(df['Age'], bins=10, color='pink', kde=False)  
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender',data=df)  
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
file2='train.csv'
df2=pd.read_csv(file2)
df2.head()
```

```python
from scipy import stats #detecting outliers
import numpy as np
df3 = df2[(np.abs(stats.zscore(df2['Fare'])) < 3)]
df3
```

```python
print(df2['Age'].hist())
```

```python
import seaborn as sns
sns.heatmap(df2.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
numeric=['Age','Fare']
df2[numeric].hist(bins=20,figsize=(10,8),color='red')
plt.suptitle("Histograms for numeric")
plt.show()
```

```python
import seaborn as sns
categoric=['SibSp','Sex','Embarked']
for feature in categoric:
    sns.countplot(x=feature,data=df2,palette='Set2',hue='Survived')
    plt.title("Distribution of categorical variables")
    plt.show()
```

```python
for feature in categoric:
    sns.countplot(x=feature,data=df2,hue='Survived')
    plt.title(f'Survival data on {feature}')
    plt.show()
```

```python
numeric=['Age','Fare']
for feature in numeric:
    sns.boxplot(x='Survived', y=feature, data=df2)
    plt.title(f'Survival Rate by {feature}')
    plt.show()
```

```python
sns.pairplot(df2, hue='Survived', diag_kind='kde')
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```
